# Exploratory Data Analysis of Superstore Sales Dataset
This notebook analyzes retail sales data from 2015 to 2018 to identify trends and insights.

```python
import pandas as pd
data = pd.read_csv('/content/sample_data/train.csv')
data.head()  # Shows the first 5 rows
```

```python
data.isnull().sum()
```

```python
data['Sales'].fillna(data['Sales'].median(), inplace=True)
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
```

```python
data.head()
```

```python
invalid_dates = data['Order Date'].isna().sum()
print(f"Number of invalid dates: {invalid_dates}")
```

```python
invalid_rows = data[data['Order Date'].isna()]
print(invalid_rows[['Order ID', 'Order Date']])
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum()
print(sales_by_category)
```

```python
import matplotlib.pyplot as plt
sales_by_category.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()  # Ensures the plot fits on your mobile screen
plt.show()
```

```python
data['Year'] = data['Order Date'].dt.year
sales_by_year = data.groupby('Year')['Sales'].sum()
print(sales_by_year)
```

```python
sales_by_year.plot(kind='line')
plt.title('Sales Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
data['Month'] = data['Order Date'].dt.month
sales_by_month = data.groupby(['Year', 'Month'])['Sales'].sum()
sales_by_month.plot(kind='line')
plt.title('Monthly Sales Trend')
plt.xlabel('Year, Month')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

# Sales Trend Analysis
The sales trend shows a decline from approximately 180,000 in 2015 to a low around 2016, followed by a significant increase through 2017 and 2018, where sales stabilized around 240,000-260,000. This could indicate a recovery phase or improved sales strategies after an initial dip. Further analysis by category or region could reveal the drivers of this growth.

```python
sales_by_year_category = data.groupby(['Year', 'Category'])['Sales'].sum()
sales_by_year_category.unstack().plot(kind='bar', stacked=True)
plt.title('Sales by Category Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

# Conclusion
The analysis reveals a sales recovery from 2016 to 2018, stabilizing at 240,000-260,000. Further investigation into monthly trends and category performance could provide deeper insights into business strategies.
